```python
# 1) Write a function to convert temperature from kelvin to celsius and celsius to kelvin
def kelvin_to_celsius(k):
    return k - 273.15

def celsius_to_kelvin(c):
    return c + 273.15
```

```python
# 2) Write a function to convert temperature to fahrenheit. 
    # Include an optional keyword argument to specify whether the input is in celcius or kelvin.
    # Call your previously defined functions if necessary.
def to_fahrenheit(temp, *, input_unit="celsius"):
    """
    Convert a temperature to Fahrenheit.
    input_unit: "celsius" or "kelvin"
    """
    unit = input_unit.lower()
    if unit == "kelvin":
        c = kelvin_to_celsius(temp)
    elif unit == "celsius":
        c = temp
    else:
        raise ValueError("input_unit must be 'celsius' or 'kelvin'")
    
    return (9/5) * c + 32
```

```python
# 3) Check that the outputs are sensible by trying a few examples
print("0°C ->", to_fahrenheit(0, input_unit="celsius"), "°F")          # should be 32°F
print("273.15 K ->", to_fahrenheit(273.15, input_unit="kelvin"), "°F") # should be 32°F
print("300 K ->", to_fahrenheit(300, input_unit="kelvin"), "°F")       # ~80.33°F
```

```python
# 4) Now write a function that converts from farenheit
    # and uses a keyword argument to specify whether you want the output in celcius or kelvin
def from_fahrenheit(f, *, output_unit="celsius"):
    """
    Convert Fahrenheit to Celsius or Kelvin.
    output_unit: "celsius" or "kelvin"
    """
    c = (f - 32) * (5/9)

    unit = output_unit.lower()
    if unit == "celsius":
        return c
    elif unit == "kelvin":
        return celsius_to_kelvin(c)
    else:
        raise ValueError("output_unit must be 'celsius' or 'kelvin'")
```

```python
# 5) Write a function that takes two arguments (feet and inches) and returns height in meters
    #Verify it gives sensible answers
def height_to_meters(feet, inches):
    total_inches = 12 * feet + inches
    meters = total_inches * 0.0254
    return meters
```

```python
print("5 ft 0 in ->", height_to_meters(5, 0), "m")   # 1.524 m
print("6 ft 0 in ->", height_to_meters(6, 0), "m")   # 1.8288 m
```

```python
# 6) Write a function takes one argument (height in meters) and returns two arguments (feet and inches)
    #(Consult the tutorial on numbers if you are stuck on how to implement this.)
def meters_to_height(meters):
    total_inches = meters / 0.0254
    feet = int(total_inches // 12)
    inches = total_inches - 12 * feet
    return feet, inches
```

```python
# 7) Verify that the “round trip” conversion from and back to meters is consistent
    # Check for 3 different values of height in meters
test_values_m = [1.50, 1.75, 2.00]

for m in test_values_m:
    ft, inch = meters_to_height(m)
    m_back = height_to_meters(ft, inch)
    print(f"Start: {m} m -> {ft} ft, {inch:.4f} in -> {m_back:.6f} m")
```
