```python
import requests
import json
```

```python
# Here's a sample JSON structure similar to what APIs return
sample_json = '''
{
  "station": "USC00305800",
  "name": "New York Central Park",
  "location": {
    "latitude": 40.7789,
    "longitude": -73.9692
  },
  "observations": [
    {"date": "2023-01-01", "temperature": 32, "precipitation": 0.0},
    {"date": "2023-01-02", "temperature": 28, "precipitation": 0.5},
    {"date": "2023-01-03", "temperature": 35, "precipitation": 0.0},
    {"date": "2023-01-04", "temperature": 38, "precipitation": 0.2},
    {"date": "2023-01-05", "temperature": 41, "precipitation": 0.0}
  ]
}
'''
```

```python
# Parse the JSON
data = json.loads(sample_json)

# Access nested data
print("Station:", data['station'])
print("Location:", data['location'])
print("First observation:", data['observations'][0])
```

```python
#PART 1
# 1. Extract and print all dates and temperatures (8 points)
print("Date, Temperature")
for obs in data['observations']:
    print(obs['date'],obs['temperature'])

#2. Calcualte average temperature (8 points)
total_temp = 0
count = 0
for obs in data['observations']:
    total_temp += obs['temperature']
    count += 1

avg_temp=total_temp/count
print(f"Average temperature: {avg_temp} degrees")

#3. Find days with precipitation (9 points)
print("/nDays with precipitation:")
for obs in data['observations']:
    if obs['precipitation']>0:
        print(obs['date'])
```

```python
import requests

def get_weather(lat, lon):
    url = f"https://api.open-meteo.com/v1/forecast?latitude={lat}&longitude={lon}&current_weather=true"
    response = requests.get(url)
    return response.json()

seoul = get_weather(37.5665, 126.9780)
print("Seoul:", seoul["timezone"])
print("Seoul weather:", seoul["current_weather"]["temperature"], "°C")

nyc = get_weather(40.7128, -74.0060)
print("NYC:", nyc["timezone"])
print("NYC weather:", nyc["current_weather"]["temperature"], "°C")

brussels = get_weather(50.8503, 4.3517)
print("Brussels:", brussels["timezone"])
print("Brussels weather:", brussels["current_weather"]["temperature"], "°C")
```
